```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
df = sns.load_dataset('tips')
print(df.head())
# instead of .relplot could hv used .scatterplot as well 
sns.relplot(x = 'total_bill',y = 'tip',data = df, kind = 'line', hue = 'smoker')
sns.relplot(x = 'total_bill',y = 'tip',data = df, kind = 'scatter', hue = 'sex', style = 'smoker', size = 'size') # style and hue are for categorical data
```

```python
# categorical (cat) plots :

# 1) cat scatter plots :
#     striplot
#     swarmplot
# 2) cat distibution plots: 
#     box plots
#     violin plots
# 3) cat estimate plot :
#     point plots
#     bar plots
#     count plots
# 4) cat categorical plots
#     heatmap
#     clustermap
# 5) facet grid plotting
```

```python
# strip plot
df = sns.load_dataset('tips')
sns.catplot(x='day',y = 'tip', kind = 'strip', data = df, hue = 'smoker') #, jitter = 0, jitter is used to define spacing between the points at the same level
sns.catplot(x='day',y = 'tip', kind = 'swarm', data = df, size = 4)
# instead of catplot could hv directly used swarmplot() or stripplot()
```

```python
sns.boxplot(df['total_bill']) #univarite plotting
sns.catplot(data= df , x = 'day' , y = 'total_bill', kind = 'box', hue = 'sex') #multivariate plotting
# boxplot shows the min and max values and the box represents normal distribution (pdf)... the dots represents outliers which shows large deviation 
```

```python
# violin plot are combination of box plot and kernel density plot (kdp)
sns.violinplot(df['total_bill'])
sns.catplot(data= df , x = 'day' , y = 'total_bill', kind = 'violin')
```

```python
# bar plot gives info abt central tendency rather than distributin of data
sns.catplot(x = 'sex', y = 'tip', data = df, kind = 'bar',estimator = np.std) #estimators can be mean,median,var,std.......default is set to mean
# black lines = error bars, it shows us confidence interval to calculate mean for the population
```

```python
sns.catplot(x = 'sex', kind = 'count',data =df)
# histogram is for numerical data, similarly countplot is like histogram for categorical data
```

```python
df1 = sns.load_dataset('flights')
print(df1.head())
x = df1.pivot_table(index = 'year', columns = 'month', values = 'passengers', aggfunc='sum')
print('\npivot table : \n',x)
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False , linewidths=0.7 , annot= True , fmt='d', cmap='viridis') 
# annotations gives a value in heatmap for better understanding, fmt is format like int/float...cmap is for color = summer /viridis /winter
```

```python
import scipy
sns.clustermap(x, metric = 'correlation', row_cluster=False) #metric is for distance calculation, standard_scale is for normalization
# decides if clustering is done on basis of rows or columns
```

```python
# jointplot :
#  scatter plot
# hexbin plot
# kde plot
# reg plot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde') # contour plots , kinda shows 2D function of pdf sparse ones are low on pdf,dense r high
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
# pair plot ,plots multiple pairwise bivariate distribution , on diagonals it compares itself with same elemnt therefor plots histogram or pdf ,other are comparison with other so scatter plot
df3 = sns.load_dataset('iris')
print(df3.head())
sns.pairplot(df3)
sns.pairplot(df3, hue = 'species')
```

```python
df4 = sns.load_dataset('titanic')
print(df4)
# univariate analysis on continous numerical data :- distplot (histograms + kernel density estimation + rugplot)
sns.distplot(df4['age'], hist = False , rug = True)
sns.distplot(df4['age'])
```
