```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random
# plt.xkcd()  #comic cstyle plotting
print(plt.style.available)            # can be used to change the style from other inbuilt types in matplotlib
plt.style.use('fivethirtyeight')
x = np.arange(25,48,2)
a = np.random.randint(25019, 44913, 12)
b = np.linspace(18824, 39132, 12)

# format stings (fmt) = '[marker][line][color]'
plt.plot(x,a,'r--', label='A')  #using fmt
# or could hv been written as plt.plot(x,a, label='A', color='r', linestyle='--')
plt.plot(x,b, label='B', color='#5a7D9A', marker='o', markersize=8, markerfacecolor='red')
plt.fill_between(x,a,b,where=(a>b),alpha = 0.28,label = 'A>B',interpolate=True)
plt.fill_between(x,a,b,where=(a<b),alpha = 0.28,label = 'A<B',interpolate=True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.grid(True)
# plt.savefig('frst_plot.png')
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
x = np.arange(25,48,2)
x_indexes = np.arange(len(x))
width  = 0.25
a = np.linspace(25019, 44913, 12)
b = np.linspace(18824, 39132, 12)
plt.bar(x_indexes,a, label='A', color='r',width = width)
plt.bar(x_indexes + width ,b, label='B', color='g',width = width)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.legend()
plt.grid(True)
plt.xticks(ticks=x_indexes, labels=x)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
matches_df = pd.read_csv('matches.csv')
deliveries_df = pd.read_csv('deliveries.csv')

merged_df = deliveries_df.merge(matches_df, left_on='match_id', right_on='id')

sixes_df = merged_df[merged_df['batsman_runs'] == 6]
sixes_count = sixes_df['batsman'].value_counts().head(20)

sixes_count.plot(kind='barh', color='skyblue')
plt.xlabel('-----------------Number of Sixes-------------->')
plt.ylabel('-----------------Player------------->')
plt.title('Top 20 Players with the Most Number of Sixes')
plt.gca().invert_yaxis()
plt.show()
```

```python
from matplotlib import pyplot as plt
slice = [120,80,30,20]
activities = ['sleeping','eating','working','playing']
plt.pie(slice, labels = activities, startangle=20, shadow=True, explode=[0,0.2,0,0], autopct='%2.2f%%')
# %x.yf%% is used where y= no of decimal places, x= no of digits before decimal/basically here spacing
plt.title('Pie Chart')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as pyplot
import random 

minutes = np.arange(1,10)
player1 = np.random.randint(1,5,9)
player2 = np.random.randint(1,4,9)
player3 = np.random.randint(1,3,9)

labels = ['player1','player2','player3']

plt.stackplot(minutes,player1,player2,player3,labels=labels)
plt.title('Stack Plot')
plt.legend(loc = 'upper right')
plt.xlabel('--------minutes played----->')
plt.ylabel('---------players-------->')
plt.show()
```
